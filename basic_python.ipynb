Lesson 4: Collection of data (list, turple, dictionary)

```python
# 1. Given numbers
numbers = [10, 20, 30, 40, 50, 60, 70, 80]

# a. Print element at index 2
print(f"Element at index 2: {numbers[2]}") 

# b. Print total element of the numbers
print(f"Total elements: {len(numbers)}") 

# c. Find last index of array
print(f"Last index: {len(numbers) - 1}") 

# d. Slicing number from index 2 to 5
print(f"Slicing index 2 to 5: {numbers[2:6]}") 

# 2. Given dict
person = {"name": "Alice", "age": 20}

# a. Print value of name
print(f"Name: {person['name']}")

# b. Add country as “Cambodia”
person["country"] = "Cambodia"
print(f"Updated person: {person}")

# 3. Given list of person
persons = [{"name": "Alice", "age": 20}, {"name": "Jonh", "age": 21}]

# a. Update Jonh’s age to 25
persons[1]["age"] = 25

# b. Add gender keys
persons[0]["gender"] = "Female"     
persons[1]["gender"] = "Male"       
print(f"Updated persons list: {persons}")
```

Lesson 5: Loop

```python
numbers = [100, 2, 3, 4, 5]

# Calculations
print(f"Min: {min(numbers)}")
print(f"Max: {max(numbers)}")
print(f"Sum: {sum(numbers)}")
print(f"Average: {sum(numbers) / len(numbers)}")

# Search index of value (e.g., 100)
search_val = 100
found = False
for i in range(len(numbers)):
    if numbers[i] == search_val:
        print(f"Index of {search_val}: {i}")
        found = True
        break
if not found:
    print("Not found")

# Print Shapes
print("\nRectangle:")
for i in range(3):
    print("* * * * *")

print("\nTriangle:")
for i in range(1, 6):
    print("* " * i)
```

Lesson 6: Functions

```python

movies = [
    {"title": "The Shawshank Redemption", 
    "director": "Frank Darabont", 
    "year": 1994, 
    "genre": ["Drama"]},
    {"title": "The Godfather", 
    "director": "Francis Ford Coppola", 
    "year": 1972, 
    "genre": ["Crime", "Drama"]}
]

# Add a new movie
def add_movie(collection, title, director, year, genres):
    new_movie = {"title": title, "director": director, "year": year, "genre": genres}
    collection.append(new_movie)

# Search for movies by title or director
def search_movies(collection, query):
    results = [m for m in collection if query.lower() in m['title'].lower() or query.lower() in m['director'].lower()]
    return results

# Filter movies based on genre or year
def filter_movies(collection, genre=None, year=None):
    results = collection
    if genre:
        results = [m for m in results if genre.lower() in [g.lower() for g in m['genre']]]
    if year:
        results = [m for m in results if m['year'] == year]
    return results

# Display movie information
def display_movies(collection):
    for m in collection:
        print(f"Title: {m['title']} | Director: {m['director']} | Year: {m['year']} | Genre: {', '.join(m['genre'])}")


# Testing functions
add_movie(movies, "Inception", "Christopher Nolan", 2010, ["Sci-Fi", "Action"])
print("All Movies:")
display_movies(movies)
print("Search Movies:")
display_movies(search_movies(movies, "Inception"))
print("Filter Movies:")
display_movies(filter_movies(movies, genre="Drama"))
```
