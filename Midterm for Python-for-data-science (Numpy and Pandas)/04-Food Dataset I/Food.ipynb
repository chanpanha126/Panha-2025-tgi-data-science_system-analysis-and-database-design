<i>Example 1. Import the necessary libraries</i>

```python
import pandas as pd
```

<i>Example 2. Import the dataset from the local file </i>

```python
order_list = pd.read_csv('./order_list.csv', sep = '\t')
```

<b>Exercise 1. Display the first 5 orders of the dataset.</b>

```python
order_list.head(5)
```

<b>Exercise 2. What is the number of rows in the dataset?</b>

```python
num_rows = len(order_list)
print(f"Number of rows: {num_rows}")
```

<b>Exercise 3. What is the number of columns in the dataset?</b>

```python
num_columns = len(order_list.columns)
print(f"Number of columns: {num_columns}")
```

<b>Exercise 4. Display the name of each column.</b>

```python
print("Column names:")
print(order_list.columns.tolist())
```

<b>Exercise 5. Display the <i>"food"</i> which has the highest number of orders?

```python
most_ordered_food = order_list['food'].value_counts().idxmax()
print(f"Most ordered food: {most_ordered_food}")
```

<b>Exercise 5. How many items were ordered for the most-ordered item?

```python
max_orders = order_list['food'].value_counts().max()
print(f"Number of times the most-ordered item was ordered: {max_orders}")
```

<b>Exercise 6. Display the <i>"description"</i> which has the highest number of orders?</b>

```python
display(order_list['description'].value_counts().head(1))
```

<b>Exercise 7. How many foods ordered in total?</b>

```python
total_foods = order_list['quantity'].sum()
print(f"Total foods ordered: {total_foods}")
```

<b>Exercise 8. Calculate the total income for the period in the dataset?</b>

```python
# clean $ sign
order_list['unit_price_clean'] = order_list['unit_price'].str.replace('$', '').astype(float)
# Calculate total income
total_income = order_list['unit_price_clean'].sum()
print(f"Total income: ${total_income:.2f}")
```

<b>Exercise 9. Find the total number of orders.</b>

```python
total_orders = order_list['id'].nunique()
print(f"Total number of orders: {total_orders}")
```

<b>Exercise 10. Find the average income for each order?</b>

```python
avg_income_per_order = order_list.groupby('id')['unit_price_clean'].sum().mean()
print(f"Average income per order: ${avg_income_per_order:.2f}")
```

<b>Exercise 11. How many different foods are ordered?</b>

```python
unique_foods = order_list['food'].nunique()
print(f"Number of different foods ordered: {unique_foods}")
```
